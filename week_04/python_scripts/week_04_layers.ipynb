## Scatter plot with Seaborn

If we want to make a plot or a graphic with subplots/facets associated with a variable in the data set, we **must** use the figure level functions in Seaborn.

A scatter plot is a kind of **relationship** plot in Seaborn, so the figure level function is `sns.relplot()`.

```python
sns.relplot(data = iris, x='Sepal.Length', y='Sepal.Width', 
            kind = 'scatter')

plt.show()
```

We want the color to denote Petal.Length, the size to denote Petal.Width and the facet to denote Species. Seaborn allows **semantic variables** hue, shape, size (in addition to x and y).

```python
sns.relplot(data = iris, x='Sepal.Length', y = 'Sepal.Width',
            hue = 'Petal.Length',
            col = 'Species',
            size = 'Petal.Width',
            kind='scatter')

plt.show()
```

```python
sns.relplot(data = iris, x='Sepal.Length', y = 'Sepal.Width',
            hue = 'Petal.Length',
            col = 'Species',
            col_wrap=1,
            size = 'Petal.Width',
            kind='scatter')

plt.show()
```

We can modify the color **palette** in Seaborn with the `palette` argument.

## Trend lines

Seaborn calls trends "regression models". There are two functions:

* figure level: `sns.lmplot()`
* axes level: `sns.regplot()`

We need to use `sns.lmplot()` to apply a trend "smoother" for each value of `Species`.

```python
sns.set_style('whitegrid')

sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           height = 8, aspect=1.7)

plt.show()
```

Specify the markers to have different shapes, we need to provide the shapes to use.

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           height = 8, aspect=1.7)

plt.show()
```

We can increase the marker size via the `scatter_kws` argument.

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter_kws={'s': 200},
           height = 8, aspect=1.7)

plt.show()
```

Can change the categorical color palette with the `palette` argument.

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter_kws={'s': 200},
           palette=sns.color_palette('dark', as_cmap=True),
           height = 8, aspect=1.7)

plt.show()
```

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter_kws={'s': 200},
           palette=sns.color_palette('colorblind', as_cmap=True),
           height = 8, aspect=1.7)

plt.show()
```

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter_kws={'s': 200},
           palette=sns.color_palette('muted', as_cmap=True),
           height = 8, aspect=1.7)

plt.show()
```

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter_kws={'s': 200},
           palette=sns.color_palette('pastel', as_cmap=True),
           height = 8, aspect=1.7)

plt.show()
```

Or we can provide manual colors.

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter_kws={'s': 200},
           palette=['c', 'y', 'm'],
           height = 8, aspect=1.7)

plt.show()
```

Turn off scatter.

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           scatter=False,
           scatter_kws={'s': 200},
           palette=['c', 'y', 'm'],
           height = 8, aspect=1.7)

plt.show()
```

Remove the confidence interval, `ci=None`.

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           ci=None,
           scatter_kws={'s': 200},
           palette=['c', 'y', 'm'],
           height = 8, aspect=1.7)

plt.show()
```

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           ci=68,
           scatter_kws={'s': 200},
           palette=['c', 'y', 'm'],
           height = 8, aspect=1.7)

plt.show()
```

```python
sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           ci=95,
           scatter_kws={'s': 200},
           palette=['c', 'y', 'm'],
           height = 8, aspect=1.7)

plt.show()
```

We need to combine the figure level and axes level functions from Seaborn to plot the regression line across all data and per group.

```python
g = sns.lmplot(data = iris, x='Sepal.Length', y='Sepal.Width',
           hue='Species',
           markers=['o', '^', 's'],
           ci=95,
           scatter_kws={'s': 200},
           palette=['c', 'y', 'm'],
           height = 8, aspect=1.7)

sns.regplot(ax = g.ax,
            data = iris, x='Sepal.Length', y='Sepal.Width',
            ci = 95,
            scatter=False)

plt.show()
```

## Group and aggregate

## Bar charts in Seaborn

Bar charts are known as countplots in Seaborn. The figure level function is `sns.catplot()` and the axes level function is `sns.countplot()`

```python
sns.catplot(data = iris, x='Species', kind='count')

plt.show()
```

Alternatively we can use the axes level function.

```python
fig, ax = plt.subplots(figsize=(12,8))

sns.countplot(data = iris, x='Species', ax=ax)

plt.show()
```

To print the number of rows on top of the bar, I base my approach on:

https://stackoverflow.com/questions/55104819/display-count-on-top-of-seaborn-barplot

```python
fig, ax = plt.subplots(figsize=(12,8))

g = sns.countplot(data = iris, x='Species', ax=ax)
```

```python
g.patches
```

```python
fig, ax = plt.subplots(figsize=(12,8))

g = sns.countplot(data = iris, x='Species', ax=ax)

for p in g.patches:
    height = p.get_height()
    g.text(p.get_x() + p.get_width()/2.,
           height + 0.1,
           height,
           ha='center')

plt.show()
```

## Heat maps

We must count first before creating the heat map. 

We can use `pd.crosstab()` to count the number of rows per combinations which is referred to as the **cross-tabulation**.  

```python
iris_copy = iris.copy()
```

```python
iris_copy['length_bin'] = iris_copy['Sepal.Length'] > iris_copy['Sepal.Length'].mean()
```

```python
iris_copy
```

```python
pd.crosstab(iris_copy.Species, iris_copy.length_bin)
```

```python
fix, ax = plt.subplots(figsize=(12, 6))

sns.heatmap(pd.crosstab(iris_copy.Species,
                        iris_copy.length_bin),
            ax=ax)

plt.show()
```

```python
fix, ax = plt.subplots(figsize=(12, 6))

sns.heatmap(pd.crosstab(iris_copy.Species,
                        iris_copy.length_bin),
            annot=True,
            ax=ax)

plt.show()
```

```python
fix, ax = plt.subplots(figsize=(12, 6))

sns.heatmap(pd.crosstab(iris_copy.Species,
                        iris_copy.length_bin),
            annot=True, annot_kws={'size': 25},
            ax=ax)

plt.show()
```

```python
fix, ax = plt.subplots(figsize=(12, 6))

sns.heatmap(pd.crosstab(iris_copy.Species,
                        iris_copy.length_bin),
            annot=True, annot_kws={'size': 25}, cbar=False,
            ax=ax)

plt.show()
```

```python
fix, ax = plt.subplots(figsize=(12, 6))

sns.heatmap(pd.crosstab(iris_copy.Species,
                        iris_copy.length_bin),
            annot=True, annot_kws={'size': 25}, cbar=False,
            linewidths=0.75, linecolor='k',
            ax=ax)

plt.show()
```
