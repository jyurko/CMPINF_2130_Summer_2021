## Visualize hwy distribution

Histogram in Seaborn is created with the figure level function `sns.displot()`.

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5)

plt.show()
```

Modify the number of bins.

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5,
             bins = 21)

plt.show()
```

Modify the binwidth.

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5,
             binwidth = 3)

plt.show()
```

```python
sns.set_style('whitegrid')

sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5)

plt.show()
```

Can turn off the fill.

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5,
             fill=False)

plt.show()
```

We can include a rug by setting `rug = True`.

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5,
             rug = True, fill=False)

plt.show()
```

To overlay a kernel density estimate with a histogram we just set `kde = True`.

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5,
             fill=False, rug=True, kde=True)

plt.show()
```

```python
sns.displot( data = mpg, x='hwy', kind='hist', aspect=1.5,
             kde=True)

plt.show()
```

The eCDF is specified with `kind = 'ecdf'`.

```python
sns.displot( data = mpg, x='hwy', kind='ecdf', aspect=1.5)

plt.show()
```

Compare the distribution between groups with a histogram by assigning `hue` to a categorical.

```python
sns.displot(data = mpg, x='hwy', hue = 'drv', kind='hist', rug=True,
            aspect = 1.75)

plt.show()
```

Density plot is easier to compare the distribution between the groups.

```python
sns.displot(data = mpg, x='hwy', hue='drv', kind='kde', rug=True, aspect=1.75)

plt.show()
```

The default Seaborn `.displot()` with `hue` assigned to a cateogorical is sensitive to the group sample size!!!!

Instead, we must override the default `common_norm` argument.

```python
sns.displot(data = mpg, x='hwy', hue='drv', kind='kde', rug=True, aspect=1.75,
            common_norm=False)

plt.show()
```

Lastly, we could use the frequency polygon.

```python
sns.displot(data = mpg, x='hwy', hue='drv',
            kind='hist', element='poly', rug=True,
            stat='density',
            binwidth=3,
            aspect=1.75,
            common_norm=False,
            fill=False)

plt.show()
```

To add "layers" requires working with the axis level functions. So let's add a marginal distribution with the conditional distributions.

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.kdeplot(data = mpg, x='hwy', c='k', ax=ax)

sns.kdeplot(data = mpg, x='hwy', hue='drv', common_norm=False, ax=ax)

plt.show()
```

For the eCDF.

```python
sns.displot(data = mpg, x='hwy', hue='drv', kind='ecdf', aspect=1.75)

plt.show()
```

## Quantiles and Boxplots

```python
mpg.groupby(['drv']).hwy.describe()
```

Boxplots are associated with the `sns.catplot()` figure level function.

```python
sns.catplot(data = mpg, x='drv', y='hwy', kind='box', aspect=1.5)

plt.show()
```

Strip plot.

```python
sns.catplot(data = mpg, x='drv', y='hwy', kind='strip', aspect=1.5)

plt.show()
```

The beeswarm plot.

```python
sns.catplot(data = mpg, x='drv', y='hwy', kind='swarm', aspect=1.5)

plt.show()
```

bee swarm with the boxplot.

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.boxplot(data = mpg, x='drv', y='hwy', ax=ax, showfliers=False)

sns.swarmplot(data = mpg, x='drv', y='hwy', color='k', ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (12, 6)) 

sns.violinplot(data = mpg, x='drv', y='hwy', ax=ax)

sns.swarmplot(data = mpg, x='drv', y='hwy', color='k', ax=ax)

plt.show()
```
