### Correlation plots in Seaborn

## Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

## Read in iris data 

```python
import statsmodels.api as sm
```

```python
iris = sm.datasets.get_rdataset('iris').data
```

```python
iris.info()
```

## Correlation matrix

```python
iris.corr()
```

```python
type( iris.corr() )
```

Create the correlation plot using `sns.heatmap()`.

```python
fig, ax = plt.subplots(figsize=(9,9))

sns.heatmap(data = iris.corr())

plt.show()
```

Set the colorbar legend bounds so that the minimum is always -1 and the maximum is always +1.

```python
fig, ax = plt.subplots(figsize=(9,9))

sns.heatmap(data = iris.corr(), vmin=-1, vmax=1, center=0)

plt.show()
```

I prefer to use the `coolwarm` color palette for correlation coefficients.

```python
fig, ax = plt.subplots(figsize=(9,9))

sns.heatmap(data = iris.corr(), vmin=-1, vmax=1, center=0, cmap='coolwarm',
            annot=True)

plt.show()
```

Modify the size of the text with the `annot_kws` argument.

```python
fig, ax = plt.subplots(figsize=(9,9))

sns.heatmap(data = iris.corr(), vmin=-1, vmax=1, center=0, cmap='coolwarm',
            annot=True, annot_kws={'size':15},
            ax=ax)

plt.show()
```

### Correlation plots per group

```python
iris.groupby(['Species']).corr()
```

Use a for-loop to iterate over the grouping variable. For each group, create the correlation plot heatmap.

```python
corr_per_species = iris.groupby(['Species']).corr()
```

```python
the_species = iris.Species.unique().tolist()

print( the_species )
```

```python
fig, axs = plt.subplots(1, len(the_species), figsize=(12,6), sharex=True, sharey=True)

for ix, x in enumerate(the_species):
    sns.heatmap(data = corr_per_species.loc[x],
                vmin=-1, vmax=1, center=0,
                cmap='coolwarm', cbar=False,
                annot=True,
                ax=axs[ix])
    
    axs[ix].set_title('Species: '+x)

plt.show()
```

## Sonar  data

```python
sonar_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data'
```

```python
sonar_df = pd.read_csv( sonar_url, header=None, prefix='X' )

print( sonar_df.shape )
```

```python
sonar_df.rename(columns={'X60': 'response'}, inplace=True)
```

```python
sonar_df.response = sonar_df.response.astype('category')
```

Boxplot summaries per variable.

```python
sns.catplot(data = sonar_df, kind='box', aspect=3.5)

plt.show()
```

Correlation plot for all numeric variables.

```python
fig, ax = plt.subplots(figsize=(11, 8))

sns.heatmap(data = sonar_df.corr(), vmin=-1, vmax=1, center=0,
            cmap='coolwarm', ax=ax)

plt.show()
```
