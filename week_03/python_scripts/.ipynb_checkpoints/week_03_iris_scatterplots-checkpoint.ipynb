### Iris scatter plot aesthetics

## Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

## Import iris

```python
import statsmodels.api as sm
```

```python
iris = sm.datasets.get_rdataset('iris').data
```

```python
iris.info()
```

Calculate the correlation matrix between the Sepal variables.

```python
iris.loc[:, ['Sepal.Length', 'Sepal.Width']].copy().corr()
```

Check the counts for the unique values of `Species`.

```python
iris.Species.value_counts()
```

How does the correlation coefficient between the Sepal variables change across the value of `Species`. Or in more general terms, how does the correlation change across **groups**.

```python
iris.loc[:, ['Species', 'Sepal.Length', 'Sepal.Width']].\
groupby(['Species']).\
corr()
```

What's going on when we group?

```python
iris.loc[:, ['Species', 'Sepal.Length', 'Sepal.Width']].\
groupby(['Species'])
```

The grouped data frame allows us to SPLIT data based on variables, then APPLY functions/methods to the different groups, and then ultimately COMBINE all the results together. This is the split-apply-combine workflow. 

```python
iris.loc[:, ['Species', 'Sepal.Length', 'Sepal.Width']].\
groupby(['Species']).\
groups
```

```python
iris.loc[:, ['Species', 'Sepal.Length', 'Sepal.Width']].\
groupby(['Species']).\
groups['setosa']
```

The groupby method creates a dictionary (effectively) behind the scenes which stores the row indices associated with each group.

```python
setosa_index = iris.loc[:, ['Species', 'Sepal.Length', 'Sepal.Width']].\
groupby(['Species']).\
groups['setosa']
```

Use the integer index to subset or select the rows associated with setosa.

```python
iris.loc[ setosa_index, ['Species', 'Sepal.Length', 'Sepal.Width']]
```

```python
iris.loc[ setosa_index, ['Species', 'Sepal.Length', 'Sepal.Width']].Species.value_counts()
```

We can apply methods to the separate groups. For example let's calculate the correlation matrix associated with the setosa group.

```python
iris.loc[ setosa_index, ['Species', 'Sepal.Length', 'Sepal.Width']].corr()
```

Double check.

```python
iris.loc[:, ['Species', 'Sepal.Length', 'Sepal.Width']].\
groupby(['Species']).\
corr()
```

Let's make a graphic in Seaborn where the x and y semantic variables (the aesthetics) are mapped to the Sepal variables and we use color, shape, and facets to denote the Species.

We want to use Seaborn figure level functions to enable easily creating the facets.

```python
sns.relplot(data = iris, 
            x='Sepal.Length', y='Sepal.Width',
            hue = 'Species', style = 'Species',
            col ='Species', 
            alpha = 0.5, s=200)

plt.show()
```

The "geometric object" in Seaborn is represented essentially by the `kind` of plot.

```python
sns.relplot(data = iris, 
            x='Sepal.Length', y='Sepal.Width',
            hue = 'Species', style = 'Species',
            col ='Species', 
            kind = 'scatter',
            alpha = 0.5, s=200)

plt.show()
```

Turn off the legend in this case since the facet is associated with the Species.

```python
sns.relplot(data = iris, 
            x='Sepal.Length', y='Sepal.Width',
            hue = 'Species', style = 'Species',
            col ='Species', 
            kind = 'scatter',
            alpha = 0.5, s=200,
            legend = False)

plt.show()
```

The Seaborn analog to ggplot2 themes is the `sns.set_style()`.

```python
sns.set_style('whitegrid')

sns.relplot(data = iris, 
            x='Sepal.Length', y='Sepal.Width',
            hue = 'Species', style = 'Species',
            col ='Species', 
            kind = 'scatter',
            alpha = 0.5, s=200,
            legend = False)

plt.show()
```

```python
sns.relplot(data = iris,
           x='Sepal.Length', y='Sepal.Width',
           s=300,
           aspect=1.75)

plt.show()
```
